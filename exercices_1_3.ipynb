### Exercice 1 - Question 2C

```python
import numpy as np

rng = np.random.default_rng(42)

a, b, sigma = 5.0, 0.8, 10.0  # paramètres du seting (cf. Ex1 Q1)

def sample_data(n):
    x = rng.uniform(30, 240, size=n)
    eps = rng.normal(0, sigma, size=n)
    y = a + b * x + eps
    return x, y
# prédicteur de Bayes
def f_star(x):                       
    return a + b * x
# estimateur naïf : constante = moyenne de Y
def f_tilde(x, y_train_mean):         
    return np.full_like(x, y_train_mean)

n_train, n_test = 2000, 20000
x_train, y_train = sample_data(n_train)
x_test, y_test = sample_data(n_test)
y_train_mean = y_train.mean()

# On approche le RISQUE RÉEL R(f) = E[(f(x) - y)^2] par une moyenne de Monte-Carlo
# sur un grand jeu de test (20000 points), pas le risque empirique sur le train.
risk_star  = np.mean((f_star(x_test) - y_test) ** 2)
risk_tilde = np.mean((f_tilde(x_test, y_train_mean) - y_test) ** 2)

print(f"Risque réel estimé, f* (Bayes) : {risk_star:.3f}")
print(f"Risque réel estimé, f_tilde    : {risk_tilde:.3f}")
print(f"Risque de Bayes théorique (sigma^2) : {sigma**2:.3f}")
```

### Exercice 2 - 1C

```python
import numpy as np

rng = np.random.default_rng(0)

def sample_data(n):                 
    x = rng.normal(0, 5, size=n)
    eps = rng.exponential(scale=1.0, size=n)
    y = x + eps
    return x, y

n_test = 200_000
x_test, y_test = sample_data(n_test)

f_star_squared  = x_test + 1.0          
f_star_absolute = x_test + np.log(2)    

risk_abs_squared_pred  = np.mean(np.abs(f_star_squared - y_test))
risk_abs_absolute_pred = np.mean(np.abs(f_star_absolute - y_test))

print(f"R_absolue(f*_carrée)  ~ {risk_abs_squared_pred:.4f}  (théorie : 2/e = {2/np.e:.4f})")
print(f"R_absolue(f*_absolue) ~ {risk_abs_absolute_pred:.4f}  (théorie : ln(2) = {np.log(2):.4f})")
```

### Exercice 3 - Q8C

```python
import numpy as np

rng = np.random.default_rng(1)

n, d = 500, 10
sigma_true = 3.0

 # fixed design
X = rng.normal(size=(n, d))          
theta_star = rng.normal(size=d)

eps = rng.normal(0, sigma_true, size=n)
y = X @ theta_star + eps

# estimateur OLS, cf. eq.(5)
theta_hat = np.linalg.solve(X.T @ X, X.T @ y)   
residuals = y - X @ theta_hat

# cf. Ex3 Q7
sigma2_hat = np.sum(residuals**2) / (n - d)     

print(f"sigma^2 réel   : {sigma_true**2:.4f}")
print(f"sigma^2 estimé : {sigma2_hat:.4f}")

# Vérification  E[sigma2_hat] = sigma^2 sur de nombreuses répétitions
n_reps = 2000
estimates = np.empty(n_reps)
for i in range(n_reps):
    eps_i = rng.normal(0, sigma_true, size=n)
    y_i = X @ theta_star + eps_i
    theta_hat_i = np.linalg.solve(X.T @ X, X.T @ y_i)
    res_i = y_i - X @ theta_hat_i
    estimates[i] = np.sum(res_i**2) / (n - d)

print(f"Moyenne empirique sur {n_reps} répétitions : {estimates.mean():.4f}")
```
