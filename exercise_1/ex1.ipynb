## Question 1

On considère le cadre de régression suivant:
- Espace d'entrée : $\mathcal{X} = [30, 240]$
- Espace de sortie : $\mathcal{Y} = \mathbb{R}_+$
- Loi jointe de $(X,Y)$ : $X \sim \mathcal{U}([30,240])$ et $Y \mid X = x \sim a + bX + \epsilon$, avec $\epsilon \sim \mathcal{N}(0, \sigma^2$), $\epsilon$ indépendant de $X$, et $a, b, \sigma > 0$ des constantes fixées
- Perte quadratique $l(\hat y, y) = (\hat y - y)^2$

On a le prédicteur de Bayes $f^*(x) = ax + b$ et le risque de Bayes $R^* = \sigma^2$. Pour plus de détails, vous trouverez les démonstrations dans la section *Bayes estimator and Bayes risk* du rapport.

## Question 2

```python
import numpy as np

rng = np.random.default_rng(42)

# Constants
a, b, sigma = 5.0, 0.8, 10.0


def sample_data(n: int) -> tuple[np.ndarray, np.ndarray]:
    """Creates the setting's input and output data.

    Parameters
    ----------
    n : int
        Number of samples to create.

    Returns
    -------
    tuple[np.ndarray, np.ndarray]
        A tuple with the input data and the corresponding output data.
    """
    x = rng.uniform(30, 240, size=n)
    eps = rng.normal(0, sigma, size=n)
    y = a + b * x + eps
    return x, y


def f_star(x: np.ndarray) -> np.ndarray:
    """Bayes predictor of the setting.

    Parameters
    ----------
    x : np.ndarray
        Input data.

    Returns
    -------
    np.ndarray
        Predicted output data.
    """
    return a + b * x


def f_tilde(x: np.ndarray, y_train_mean: float) -> np.ndarray:
    """Naive Bayes prediction using mean output data.

    Parameters
    ----------
    x : np.ndarray
        Input data.
    y_train_mean : float
        Average of the output data.

    Returns
    -------
    np.ndarray
        Predicted output data, the mean each time.
    """
    return np.full_like(x, y_train_mean)


n_train, n_test = 2000, 20000
x_train, y_train = sample_data(n_train)
x_test, y_test = sample_data(n_test)
y_train_mean = y_train.mean()

# We estimate the true risk R(f) = E[(f(x) - y)^2] by Monte-Carlo mean
# on a big testing set (20000 points), not empirical mean of the training set.
risk_star = np.mean((f_star(x_test) - y_test) ** 2)
risk_tilde = np.mean((f_tilde(x_test, y_train_mean) - y_test) ** 2)

print(f"Estimated true risk, f* (Bayes)  : {risk_star:.3f}")
print(f"Estimated true risk, f_tilde     : {risk_tilde:.3f}")
print(f"Theoretical Bayes risk (sigma^2) : {sigma**2:.3f}")
```
