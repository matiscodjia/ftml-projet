## Question 1-7

Les questions 1 à 6 prouvent la proposition $\mathbb{E}\big[R_X(\hat\theta)\big] = \frac{n-d}{n}\sigma^2$. Avec la question 7, on montre qu'un estimateur de $\sigma^2$ est $\hat\sigma^2 = \dfrac{\|y-X\hat\theta\|^2}{n-d}$.

Pour plus de détails, vous trouverez les démonstrations dans la section *Expected value of empirical risk for OLS* du rapport.

## Question 8

```python
import numpy as np

rng = np.random.default_rng(1)

n, d = 500, 10
sigma_true = 3.0

# fixed design
X = rng.normal(size=(n, d))
theta_star = rng.normal(size=d)

eps = rng.normal(0, sigma_true, size=n)
y = X @ theta_star + eps

# OLS estimator, cf. eq.(5)
theta_hat = np.linalg.solve(X.T @ X, X.T @ y)
residuals = y - X @ theta_hat

# Sigma^2 estimator cf. Ex3 Q7
sigma2_hat = np.sum(residuals**2) / (n - d)

print(f"True sigma^2      : {sigma_true**2:.4f}")
print(f"Estimated sigma^2 : {sigma2_hat:.4f}")

# Verification of E[sigma2_hat] = sigma^2 over multiple iterations
n_reps = 2000
estimates = np.empty(n_reps)
for i in range(n_reps):
    eps_i = rng.normal(0, sigma_true, size=n)
    y_i = X @ theta_star + eps_i
    theta_hat_i = np.linalg.solve(X.T @ X, X.T @ y_i)
    res_i = y_i - X @ theta_hat_i
    estimates[i] = np.sum(res_i**2) / (n - d)

print(f"Empirical mean over {n_reps} iterations : {estimates.mean():.4f}")
```
