## Question 1

Le setting proposé est le suivant:
- $\mathcal{X} = \mathbb{R}$
- $\mathcal{Y} = \mathbb{R}$
- $Y \mid X = x \sim x + \epsilon$ avec $\epsilon \sim \mathcal{E}(1)$ et $\epsilon$ indépendant de X.

Pour plus de détails, vous trouverez les démonstrations dans la section *Bayes risk with absolute loss* du rapport.

```python
import numpy as np

rng = np.random.default_rng(0)


def sample_data(n: int) -> tuple[np.ndarray, np.ndarray]:
    """Draw n input/output samples from the setting."""
    x = rng.normal(0, 5, size=n)
    eps = rng.exponential(scale=1.0, size=n)
    y = x + eps
    return x, y


n_test = 200_000
x_test, y_test = sample_data(n_test)

f_star_squared = x_test + 1.0
f_star_absolute = x_test + np.log(2)

risk_abs_squared_pred = np.mean(np.abs(f_star_squared - y_test))
risk_abs_absolute_pred = np.mean(np.abs(f_star_absolute - y_test))

print(
    f"R_absolue(f*_squared) ~ {risk_abs_squared_pred:.4f}  (theoretical : 2/e = {2 / np.e:.4f})"
)
print(
    f"R_absolue(f*_absolue) ~ {risk_abs_absolute_pred:.4f}  (theoretical : ln(2) = {np.log(2):.4f})"
)
```
